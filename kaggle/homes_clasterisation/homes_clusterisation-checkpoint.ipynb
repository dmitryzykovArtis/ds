## Алгоритм кластеризации

```python
# import ma_note_bot
from array import *

def clasterisation(df, distance_border):
    
    """
    Notes
    -----
    По возможности используем Numpy т.к. он осуществляет работу с матрицами быстрее (в фортране)

    Будем использовать сортировку по долготе, и для каждого объекта
    рассматривать только те объекты расстояние по долготе в которых
    не привышает distance_border.
    
    Для простоты расстоянием до кластера будем считать расстояние до
    ближайшей точки этого кластера.
    не привышает distance_border.

    Для возможности осуществления обратной сортировки добавим колонку id
    
    Переведем DataFrame в ndarray для увеличения скорости работы, так же долготу отдельно переведем в array

    """
    
    def find_distance(x1, x2):
        return np.linalg.norm(x1[:-2]-x2[:-2])

    df.sort_values(by = 'Долгота', inplace=True)
    longitude = array('f', df['Долгота'].to_numpy())
    claster = array('L', np.zeros(len(df), dtype=int))
    df_np = df.astype('float32').to_numpy()

    claster_count = 1
    
    for i in range(1, len(df)):
        if i%10000 == 0:
            print('Номер текущей итерации:', i)
        j = i-1
        min_distance = 1
        min_j = 0
        
        while j > 0 and longitude[i] - longitude[j] < distance_border:
            distance = find_distance(df_np[i], df_np[j])
            if distance < distance_border and distance < min_distance:
                min_distance = distance
                min_j = j
            j -= 1
        
        if min_distance < 1:
            claster[i] = claster[min_j]
        else:
            claster[i] = claster_count
            claster_count += 1
    
    df['сluster'] = claster
    df.sort_index(inplace=True)
    return df
DISTANCE = 10
df_clasters = clasterisation(df_small[:][:], DISTANCE)

# использовал телеграм бота для нотификации окончания работы модели
# ma_note_bot.send_telegram("Clusterisation complite!")

# использовал профайлер для оптимизации скорости работы алгоритма
# import line_profiler
# lp = line_profiler.LineProfiler()
# lp.add_function(clasterisation)
# lp.runctx('clasterisation(df_small[:][:10000], DISTANCE)', locals=locals(), globals=globals())
# lp.print_stats()    
    
    
    
        
```

## Вывод по кластеризации

Плюсы:
Построенная кластеризация хорошо работает для нахождения объявлений с одинаковых квартир в одном доме, при этом в случае если в квартире сделали ремонт, либо с момента подачи старого объявления прошло продолжительное время, кластеризация с данными весами отнесет объявления к одному кластеру.

Минусы:
В случае наличия в доме на одном этаже в разных подъездах квартир с одинаковым числом комнат, общей площадью. То модель отнесет данные объявления к одному кластеру.

Возможные улучшения модели:
Предлагается, сделать К моделей с различными весами, возможно в некоторые модели добавить  **Дата публикации объявления** приведенную в epoch и **Состояние** трансформированое при помощи OHE (order не подходит т.к. есть параметр 'не указано') и для конкретной бизнес задачи при помощи алгоритмов многоруких бандитов подобрать модель с весами максимизирующими целевую метрику.
Предлагается для различения одинаковых квартир на одном этаже использовать фотографии приложенные к объявлению, если данная информация присутствует. Т.к. в случае повторной подачи объявления через некоторое время пользователь скорее всего будет использовать одни и те же фотографии. А разные пользователи из одного и того же дома - разные.
Так же возможно использование текста объявления с парсингом по словам/буквам, но в данном случае надо рассматривать полезность отдельно. Т.к. если пользователь переподает объявление, возможно текст объявления будет иным.
